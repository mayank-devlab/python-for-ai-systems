# Args and Kwargs

*args* (short for "arguments") allows you to pass a variable number of non-keyword arguments to a function. Imagine you're hosting a dinner party and you want to invite friends. Instead of saying, "I can only invite two friends," you say, "Bring as many friends as you want!" In coding terms, you use *args to let the function accept any number of arguments. For example:

```
def sum_of(*args):
    total = 0
    for number in args:
        total += number
    return total

print(sum_of(4, 5, 6))  # This will output 15

```


*kwargs* (short for "keyword arguments") allows you to pass a variable number of keyword arguments (which are key-value pairs) to a function. Think of it like ordering food at a restaurant where you can specify what you want: "Iâ€™d like a coffee for $2.99, a cake for $4.55, and a juice for $2.99." In coding, you use **kwargs to accept any number of keyword arguments. Here's how it looks:

```
def calculate_bill(**kwargs):
    total = 0
    for item, price in kwargs.items():
        total += price
    return total
print(calculate_bill(coffee=2.99, cake=4.55, juice=2.99))  # This will output 10.53

```
