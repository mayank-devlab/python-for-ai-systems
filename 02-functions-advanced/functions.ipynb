# functions

```python
def calculate_tax(bill, tax_rate):
    """
    Calculate the tax based on bill amount and tax rate.

    Parameters:
    bill (float): The bill amount.
    tax_rate (float): The tax rate as a percentage.

    Returns:
    float: The calculated tax.
    """
    return round(bill * (tax_rate / 100), 2)

print("Total Tax due:", calculate_tax(100, 5))  # Example usage, should return 5.0


    
```
