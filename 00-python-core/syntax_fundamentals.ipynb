# match case

```python
"""Match case is introduce from python 3.10 version onwards
and its the alternative of if-elif-else ladder
"""

https_status = 404

match https_status:
    case 200:
        print("OK")
    case 404| 403:
        print("Not Found")
    case 500:
        print("Internal Server Error")
    case _:
        print("Unknown Status")
```

# enumerate

```python
favourite = ['creme brulee', 'tiramisu', 'cheesecake', 'chocolate mousse', 'apple pie']

for idx, dessert in enumerate(favourite):
    print("I like this dessert number:", idx, ":", dessert)
```
